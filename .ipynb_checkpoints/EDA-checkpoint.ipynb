```python
#3 Reltionship analysis
```

```python
plt.figure(figsize=(10,10))
ax = sns.countplot(x=data["class"],  data=data, order = data["class"].value_counts().index )
for p, label in zip(ax.patches, data["class"].value_counts()):   
    ax.annotate(label, (p.get_x()+0.25, p.get_height()+0.5))
```
